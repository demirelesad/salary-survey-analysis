# Doğrusal Programlama Örnek

Amaç Fonksiyonu 

$max_z$ = $3x_1 + x_2$

Kısıtlar

$x_2 \le 5$

$x_1 + x_2 \le 10$

$x_1 - x_2 \le 2$

$x_1, x_2 \ge 0$

```python
import numpy as np
```

```python
from scipy.optimize import linprog
```

```python
A = np.array([[0,1],[1,1],[1,-1]]) #x1 ve x2 katsayilari
B = np.array([5,10,2]) #kisitlarin maksimum degeri
C = np.array([-3,-1]) #amac fonksiyonu - koymamizin nedeni resultta amacimizin maks oldugunu soylemememiz. - koyarak ceviriyoruz.
```

```python
linprog(c = C, A_ub = A, b_ub = B, method = "simplex", bounds = (0,None))
```

```python
result = linprog(c = C, A_ub = A, b_ub = B, method = "simplex", bounds = (0,None))
```

```python
result.fun # maks (+22)
```

```python
result.x # x1 ve x2
```

- birinci grafik = $y = 5 - 0*x_1$
- ikinci grafik = $y = 10 - x_1$
- üçüncü grafik = $y = x_1 - 2$ 
- amaç fonksiyonu grafiği = $y = 22 - 3x_1$

```python
import matplotlib.pyplot as plt 
```

```python
X = np.linspace(0,10,10) #0 ile 10 arasinda 10 adet esit aralikli sayi uretme
y1 = 5 - 0*X
y2 = 10 - X
y3 = X - 2
y4 = 22 - 3*X
```

```python
plt.plot(X, y1, label = "$y = 5 - 0*x_1$")
plt.plot(X, y2, label = "$y = 10 - x_1$")
plt.plot(X, y3, label = "$y = x_1 - 2$")
plt.plot(X, y4, label = "$Z = 3x_1 + x_2$", linestyle = "dashed")
plt.legend()
plt.scatter(result.x[0],result.x[1])
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.xlim(left = 0)
plt.ylim(bottom = 0, top = 10)
plt.style.use("classic")
```
